```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# Memuat dataset
url = "https://drive.google.com/uc?id=1QuDHd-pebfOY-DoyFACtMyK-_UyuMJP8"
data = pd.read_csv(url)
print(data.head())
```

```python
# Memeriksa tipe data dan kolom
print(data.info())
```

```python
# Menampilkan nama kolom
print(data.columns.tolist())
```

```python
# Normalisasi data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[['x', 'y']]) 
```

```python
# KMeans Clustering
kmeans = KMeans(n_clusters=5, random_state=42)
data['KMeans_Cluster'] = kmeans.fit_predict(data_scaled)

# Visualisasi hasil KMeans
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['x'], y=data['y'], hue=data['KMeans_Cluster'], palette='viridis')
plt.title('KMeans Clustering')
plt.show()
```

```python
# Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=5)
data['Hierarchical_Cluster'] = hierarchical.fit_predict(data_scaled)

# Visualisasi hasil Hierarchical Clustering
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['x'], y=data['y'], hue=data['Hierarchical_Cluster'], palette='viridis')
plt.title('Hierarchical Clustering')
plt.show()
```

```python
# DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
data['DBSCAN_Cluster'] = dbscan.fit_predict(data_scaled)

# Visualisasi hasil DBSCAN
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['x'], y=data['y'], hue=data['DBSCAN_Cluster'], palette='viridis')
plt.title('DBSCAN Clustering')
plt.show()
```
